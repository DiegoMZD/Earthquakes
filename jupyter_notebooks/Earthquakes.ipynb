3 Function Premonitory Tremors  
4 Filter

```python
from geopy.distance import geodesic
from datetime import timedelta
def calculate_distance(center, point):
    return geodesic(center, point, ellipsoid="WGS-84").km

def premonitory_tremors(earthquake, tremors_list, radius_km, period_days):
    center = (earthquake.latitude, earthquake.longitude)

    # Filter by location
    filter_radius = tremors_list.apply(
        lambda tremor: calculate_distance(center, (tremor.latitude, tremor.longitude)) <= radius_km, axis=1
    )

    # Filter by time
    initial_time = earthquake.date - timedelta(period_days)
    final_time = earthquake.date
    filter_time = (tremors_list['date'] >= initial_time) & (tremors_list['date'] < final_time)

    premonitory_tremors_list = tremors_list[filter_radius & filter_time]
    return premonitory_tremors_list
```

```python
#new dataframe
earthquakes_and_premonitory = earthquakes_list.drop(columns=['tsunami','data_type'])
from statistics import mean

def avg_and_number_premonitory_tremors(earthquakes_list, tremors_list, radius_km, period_days):
    premt_avg_mag = []
    premt_quantity = []

    for _, earthquake in earthquakes_list.iterrows():
        premonitory_tremors_list = premonitory_tremors(earthquake, tremors_list, radius_km, period_days)
        premt_quantity_e = len(premonitory_tremors_list)
        premt_avg_mag_e = premonitory_tremors_list['magnitudo'].mean() if premt_quantity_e > 0 else 0
        premt_avg_mag.append(premt_avg_mag_e)
        premt_quantity.append(premt_quantity_e)

    return premt_avg_mag, premt_quantity
    
period_days_list = [1,2,3,4,5,6,7]
radius_km = 250
for period_days in period_days_list:
    result = avg_and_number_premonitory_tremors(earthquakes_list, tremors_list, radius_km, period_days)
    avg_column_name = f"Avg_Magnitude_{period_days}days"
    num_column_name = f"Number_of_Tremors_{period_days}days"
    earthquakes_and_premonitory[avg_column_name] = result[0]
    earthquakes_and_premonitory[num_column_name] = result[1]
    print("Day {} done".format(period_days))

condition = earthquakes_and_premonitory.date >= firts_date_tremor + timedelta(max(period_days_list))
earthquakes_and_premonitory = earthquakes_and_premonitory[condition]
earthquakes_and_premonitory.to_csv(r'..\data\earthquakes_and_premonitory.csv')
earthquakes_and_premonitory.head()
```
